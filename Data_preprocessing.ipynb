```python
#find the missing values 
dft.isnull().sum(axis = 0)
```

```python
#data preproesiing replacing the missing values 
```

```python
dt = dft.copy()
```

```python
#replace the values with mode 
cols = ['Gender','Area','Loan_Status']
dt[cols] = dt[cols].fillna(dt.mode().iloc[0])
```

```python
dt.isnull().sum(axis = 0)
```

```python
#replacing the numerical value 
cols2 = ['ApplicantIncome','CoapplicantIncome','LoanAmount']
dt[cols2] = dt[cols2].fillna(dt.mean())
```

```python
dt.isnull().sum(axis = 0)
```

```python
#Label encoding for categorical variable 
dt.dtypes
```

```python
dt[cols] = dt[cols].astype('category')
```

```python
dt.dtypes
```

```python
for columns in cols:
    dt[columns] = dt[columns].cat.codes
```
